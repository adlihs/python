```python
def team_clusters(clusters,pass_data=pass_data,team=None, game=None,player=None):
  import numpy as np
  from sklearn.cluster import KMeans
  team = team
  game = game
  player = player


  # Load team's pass data
  #if team != None and game == None:
  #  team_1_passes = pass_data.query("team == @team ")
  #elif team !=None and game != None:
  #  team_1_passes = pass_data.query("team == @team and game == @game ")
  #elif team ==None and game == None and player != None:
  #  team_1_passes = pass_data.query("player == @player ")

  team_pass = pass_data[pass_data['team'] == team]
  team_pass = team_pass[(team_pass['x'] >= 50) & (team_pass['outcome_type'] == 'Successful') &
                        (team_pass['end_x'] >= 50)]
  print(len(team_pass))
  team_pass['pass_distance'] = np.sqrt(np.square(team_pass['end_x'].astype(float) - team_pass['x'].astype(float)) + np.square(team_pass['end_y'].astype(float) - team_pass['y'].astype(float)) )


  # New pass dataframe
  #pass_team_1 = pd.DataFrame()
  #pass_team_1['x'] = team_1_passes.pos_x
  #pass_team_1['y'] = team_1_passes.pos_y
  #pass_team_1['end_x'] = team_1_passes.pos_x_end
  #pass_team_1['end_y'] = team_1_passes.pos_y_end
  #pass_team_1['player'] = team_1_passes.jugador
  #pass_team_1['recipient'] = team_1_passes.jugador_end
  #pass_team_1['pass_distance'] = np.sqrt(np.square(pass_team_1['end_x'].astype(float) - pass_team_1['start_x'].astype(float)) + np.square(pass_team_1['end_y'].astype(float) - pass_team_1['start_y'].astype(float)) )

  # Cluster definition
  N_clusters = clusters
  X_1 = np.array(team_pass[['x', 'y', 'end_x', 'end_y']])

  kmeans = KMeans(n_clusters = N_clusters, random_state = 0).fit(X_1)
  cluster_labels = kmeans.predict(X_1)
  team_pass['n_cluster'] = cluster_labels
  centroids_1 = pd.DataFrame(data = kmeans.cluster_centers_, columns = ['x', 'y', 'end_x', 'end_y'])

  return team_pass, centroids_1
```

```python
import pandas as pd
def modificar_fila_anterior(df, indice, columna1, columna2, valor1, valor2):
  """
  Modifica los valores de dos columnas en la fila previa a un índice en un DataFrame.

  Args:
      df: El DataFrame.
      indice: El índice de la fila a partir de la cual se busca la fila anterior.
      columna1: La primera columna a modificar.
      columna2: La segunda columna a modificar.
      valor1: El nuevo valor para la primera columna.
      valor2: El nuevo valor para la segunda columna.

  Returns:
      El DataFrame modificado.
  """

  fila_anterior = df.iloc[indice - 1]
  fila_anterior[columna1] = valor1
  fila_anterior[columna2] = valor2
  return df

# Ejemplo de uso
df = pd.DataFrame({'A': [1, 2, 3, 4], 'B': [5, 6, 7, 8]})

df
```
